Q10. Write to the file all prime numbers between 600 and 800.


```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Generate prime numbers between 600 and 800
prime_numbers = [num for num in range(600, 801) if is_prime(num)]

# Write prime numbers to a file
with open('prime_numbers_600_to_800.txt', 'w') as file:
    for prime in prime_numbers:
        file.write(str(prime) + '\n')

print(f"Writing done !! \nOpen prime_numbers_600_to_800.txt to view the content ")
print(f"Total prime numbers found: {len(prime_numbers)}")
```

Q12. WAP to sort following number of elements in a list, calculate time taken and plot the graph.

```python
#Number of elements in list        Time Taken

# 5k                                  T1
# 10k                                 T2
# 15k                                 T3
# 20k                                 T4
# 25k                                 T5

import time
import matplotlib.pyplot as plt

def bubble_sort(arr):
    n = len(arr)
    for i in range(n-1):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

def generate_list(size):
    return list(range(size))

def measure_time_sorting(size):
    arr = generate_list(size)
    start_time = time.time()
    bubble_sort(arr)  # You can replace with other sorting algorithms
    end_time = time.time()
    return end_time - start_time

sizes = [5000, 10000, 15000, 20000, 25000]
times = []

for size in sizes:
    time_taken = measure_time_sorting(size)
    times.append(time_taken)
    print(f"Size:{size}, Time Taken:{time_taken} seconds")

plt.plot(sizes, times)
plt.xlabel('number os elements in list')
plt.ylabel('time taken (seconds)')
plt.title('sorting time vs number of elements')
plt.show()
```
